# Classify movies based on ratings (Excellent, Good & Average)

```python
df.columns
```

```python
def rating(rating):
    if rating >= 7.0:
        return "Excellent"
    elif rating >= 6.0:
        return "Good"
    else:
        return "Average"
```

```python
df['Rating Category'] = df['Rating'].apply(rating)
```

# The unique values from Genre

```python
df.columns
```

```python
df['Genre']
```

```python
list1 = []
for value in df ['Genre']:
    list1.append(value.split(','))
```

```python
one_d = []
for item in list1:
    for item1 in item:
        one_d.append(item1)
        
one_d        
```

```python
uni_list = []
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)

uni_list
```

```python
data = {
    'Genre': ['Drama', 'Action', 'Comedy', 'Adventure', 'Thriller', 'Crime', 'Romance', 'Sci-Fi', 'Horror', 'Mystery',
              'Fantasy', 'Biography', 'Family', 'Animation', 'History', 'Sport', 'Music', 'War', 'Western', 'Musical'],
    'Count': [513, 303, 279, 259, 195, 150, 141, 120, 119, 106, 101, 81, 51, 49, 29, 18, 16, 13, 7, 5]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 8))
sns.barplot(x='Count', y='Genre', data=df, orient='horizontal')
plt.title('Count of Movies by Genre')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

The chart suggests that Drama movies are the most preferred genre among viewers.
